```python
import sqlite3
import pandas as pd

sql_table_names = """
              SELECT name 
              FROM sqlite_master 
              WHERE type='table'
              """

sql_tuition = """
              select *
              from ipeds_raw_tuition
              """

sql_directory = """
              select *
              from ipeds_raw_tuition
              """
sql_cips = """
               select *
               from ipeds_raw_fips_file
"""

sql = "drop table ipeds_raw_cips_file"

with sqlite3.connect("../ipeds.db") as con:
    df_tables = pd.read_sql(sql_table_names,con)
    df_raw_tuition = pd.read_sql(sql_tuition,con)
    df_raw_directory = pd.read_sql(sql_directory,con)
    df_raw_fips = pd.read_sql(sql_cips,con)

df_tables
```

```python
df_raw_tuition.to_csv('tuition.csv')
df_raw_directory.to_csv('directory.csv')
df_raw_fips.to_csv('cips.csv')
```
